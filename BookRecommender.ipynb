## Popularity based recommendation system

```python
ratings.merge(books, on='ISBN')
```

```python
ratings.head(2)
```

```python
# not all books in ratings df are available in books df
ratings[ratings['ISBN'].isin(books['ISBN'])]
```

```python
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
ratings_with_name.head()
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].count().reset_index()
```

```python
num_rating_df.head()
```

```python
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
```

```python
num_rating_df.head()
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'}, inplace=True)
avg_rating_df.head(2)
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df.head()
```

```python
popular_df = popular_df[popular_df['num_ratings'] >= 250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popular_df.head(2)
```

```python
popular_df.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')
```

```python
popular_df.merge(books, on = 'Book-Title').drop_duplicates('Book-Title').shape
```

```python
popular_df = popular_df.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_ratings', 'avg_ratings']]
```

## Collaborative Recommender

```python
ratings_with_name.head()
```

```python
x = ratings_with_name.groupby('User-ID')['Book-Rating'].count() > 200
```

```python
x[x].index
```

```python
avid_readers = x[x].index
avid_readers
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(avid_readers)]
```

```python
filtered_rating.head(2)
```

```python
y = filtered_rating.groupby('Book-Title')['Book-Rating'].count() >= 50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores 
```

```python
sorted(list(enumerate(similarity_scores[0])),key=lambda x:x[1], reverse=True)[1:6]
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    distances = similarity_scores[index]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1], reverse=True)[1:6]

    data =[]
    for i in similar_items:
        #print(pt.index[i[0]])
        item=[]
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)
        
    return data
```

```python
recommend('1984')
```
